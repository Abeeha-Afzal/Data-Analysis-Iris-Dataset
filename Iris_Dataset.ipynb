# Feature Scaling - Standardization

## By using our own formula

```python
features['SepalLengthCm'] = (features['SepalLengthCm'] - features['SepalLengthCm'].mean()) / features['SepalLengthCm'].std()
```

```python
features
```

```python
features['SepalWidthCm'] = (features['SepalWidthCm'] - features['SepalWidthCm'].mean()) / features['SepalWidthCm'].std()
```

```python
features
```

```python
features['PetalLengthCm'] = (features['PetalLengthCm'] - features['PetalLengthCm'].mean()) / features['PetalLengthCm'].std()
```
